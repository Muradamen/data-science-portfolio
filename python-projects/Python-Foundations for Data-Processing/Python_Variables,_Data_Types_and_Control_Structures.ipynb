
## 4. Sets

A **set** is an unordered, mutable collection of unique elements. Sets automatically remove duplicates and are
useful for membership tests and mathematical operations like union, intersection, and difference.

**Example:**
```python
my_set = {1, 2, 3, 3, 4}
print(my_set)   # Output: {1, 2, 3, 4}
```

**Applications:** Sets are used to eliminate duplicate entries, perform fast membership testing, and handle operations like finding common or distinct elements between datasets.

### Common Set Methods
- `add(x)` – Adds an element to the set  
- `remove(x)` – Removes x from the set (error if not present)  
- `discard(x)` – Removes x if present (no error if not)  
- `union(other)` – Returns union of sets  
- `intersection(other)` – Returns common elements  
- `difference(other)` – Returns elements not in other  
- `issubset(other)` – Checks if set is subset  
- `issuperset(other)` – Checks if set is superset  


````python
# Working with Sets
set1 = {1, 2, 3}
set2 = {3, 4, 5}
print(set1.union(set2))        # {1, 2, 3, 4, 5}
print(set1.intersection(set2)) # {3}
````


## `Match` statement

The `match` statement checks a value against multiple patterns and executes the first matching block of code.

**Syntax**

```python
match variable:
    case pattern1:
        # code block
    case pattern2:
        # code block
    case _:
        # default block (like "else")
```

- `match variable`: → the value you’re checking.

- `case pattern`: → each possible value or condition.

- `case _`: → the wildcard (matches anything, works like else).


**Example**

- Using if statement

````python
day = "Friday"

if day == "Monday":
    print("Start of the week 😩")
elif day == "Friday":
    print("Almost weekend 🎉")
elif day == "Sunday":
    print("Time to rest 😴")
else:
    print("Just another day...")
````

- Using case statement

````python
day = "Monday"

match day:
    case "Monday":
        print("Start of the week 😩")
    case "Friday":
        print("Thank God, its Friday 🎉")
    case "Sunday":
        print("Time to rest 😴")
    case _:
        print("Just another day...")
````

## **`while`** loops

A while loop in Python is used to repeat a block of code as long as a condition is true. Unlike the for loop, the number of iterations is not fixed—it depends on whether the condition remains true. If the condition becomes false, the loop stops.

**Syntax**
```python
while condition:
    # code block
```

````python
# counting
i = 1
while i <= 5:
    print(i)
    i += 1
````

````python
# break for early termination
i = 1
while True:
    print("Looping:", i)
    if i == 3:
        break
    i += 1
````

````python
# continue
i = 0
while i < 5:
    i += 1
    if i == 3:
        continue
    print(i)
````

````python
# Using break
while True:  # infinite loop
    user_input = input("Type 'exit' to stop: ")
    if user_input.lower() == "exit":
        print("Loop ended safely.")
        break
    else:
        print("You typed:", user_input)
````

**Comparison of for vs while Loops in Python**

| Feature                   | **for loop**                                                               | **while loop**                                                           |
| ------------------------- | -------------------------------------------------------------------------- | ------------------------------------------------------------------------ |
| **Definition**            | Repeats a block of code for each item in a sequence or range.              | Repeats a block of code **as long as** a condition remains true.         |
| **When to Use**           | When the number of iterations is **known** or based on a sequence.         | When the number of iterations is **unknown** and depends on a condition. |
| **Control**               | Controlled by elements of a sequence (`list`, `tuple`, `string`, `range`). | Controlled by a condition (`True`/`False`).                              |
| **Syntax**                | `for variable in sequence:`                                                | `while condition:`                                                       |
| **Example**               | `for i in range(5): print(i)`                                              | `i=0; while i<5: print(i); i+=1`                                         |
| **Risk of Infinite Loop** | Rare (since it usually iterates over a finite sequence).                   | More likely (if the condition never becomes `False`).                    |


# Functions

A function in Python is a reusable block of code that performs a specific task. Functions help make programs more organized, reduce repetition, and improve readability. They can take inputs (called parameters), process them, and optionally return an output using the return statement.

Python provides two types of functions:

1. Built-in functions – already provided by Python (e.g., print(), len(), type()).
2. User-defined functions – created by programmers using the def keyword.

**Syntax**
```python
def function_name(parameters):
    # code block
    return result
```

- **def** → keyword to define a function.

- **function_name** → name of the function (should be descriptive).

- **parameters** → optional inputs (also called arguments).
- **arguments** → value you pass to a function when calling it. Arguments give functions the data they need to work with.

  - When you define a function, you specify parameters (placeholders).
  - When you call the function, you provide arguments (actual values).
- **return** → optional, sends a value back to the caller.

### Arguments

Arguments can be passed to a function in several ways:
- **Positional Arguments**: Must be supplied in the correct order as defined by the parameters.
``` python
def function_name(arg1, arg2, ...):
  # code block
```
- **Keyword Arguments**: Arguments are identified by their parameter name, meaning their order doesn't matter.

``` python
def function_name(param1=arg1, param2=arg2, ...):
  # code block
```

- **Default Arguments** You can assign a default value to a parameter in the function definition. If the caller doesn't provide an argument for that parameter, the default value is used.

``` python
def function_name(name, message="Hello"):
  # code block
```

## Problem Statement

**White Ice** is a business that currently prepares invoices manually whenever customers purchase items. The cashier calculates item totals, applies discounts (if the customer is a member), and writes out the invoice by hand. This process is time-consuming and prone to calculation errors, especially during busy periods when there are many customers. The business needs a reliable and efficient solution to automate its invoice generation process.

````python
# ------------------------------
# Invoice Program with Lists
# Demonstrates variables, comments, conditionals, data types, functions, and lists
# ------------------------------

# Function to calculate total price
def calculate_total(prices, quantities, discount=0):
    """
    Calculate the total cost of items in the invoice
    :param prices: list of floats - prices of items
    :param quantities: list of ints - quantities of items
    :param discount: float - discount percentage
    :return: float - final total
    """
    subtotal = 0
    for i in range(len(prices)):
        subtotal += prices[i] * quantities[i]   # multiply each price with quantity

    discount_amount = subtotal * (discount / 100)
    total = subtotal - discount_amount
    return subtotal, discount_amount, total

````

````python

# --- Input Section ---
customer_name = input("Enter Customer Name: ")

# Lists to store multiple items
items = []
prices = []
quantities = []

num_items = int(input("Enter number of items purchased: "))

for i in range(num_items):
    print(f"\nItem {i+1}:")
    item_name = input("Enter Item Name: ")
    item_price = float(input("Enter Item Price: "))
    item_quantity = int(input("Enter Item Quantity: "))

    items.append(item_name)
    prices.append(item_price)
    quantities.append(item_quantity)

````

````python
# Membership discount
membership = input("\nIs the customer a member? (yes/no): ").lower()
is_member = True if membership == "yes" else False
discount_percent = 10 if is_member else 0


````

````python
# --- Processing ---
subtotal, discount_amount, final_total = calculate_total(prices, quantities, discount_percent)

````

````python
# --- Output Section: Invoice ---
print("\n====== INVOICE ======")
print("Customer:", customer_name)
print("----------------------")
print("Items Purchased:")
for i in range(num_items):
    print(f"{items[i]} - {quantities[i]} x {prices[i]} = {quantities[i] * prices[i]}")
print("----------------------")
print("Subtotal:", subtotal)
print("Discount:", discount_amount, f"({discount_percent}%)")
print("Total Amount Payable:", final_total)
print("======================")
````
