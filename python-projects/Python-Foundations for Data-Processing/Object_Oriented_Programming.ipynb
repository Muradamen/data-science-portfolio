# Introduction
Object-Oriented Programming (OOP) is a programming paradigm that organizes code around objects — reusable components that combine data (attributes) and behavior (methods).
It helps make software more modular, maintainable, and scalable.

## Key Components of OOP
- **Class:** A blueprint or a template for creating objects. It defines the structure (attributes) and behavior (methods) that all objects of that type will have.

- **Object:** A specific instance of a class. It is the actual entity created at runtime that holds data and has behavior defined by its class.

- **Method:** A function defined within a class that performs an action. It represents the behavior of an object.

- **Attribute (or Field):** A variable defined within a class that holds data. It represents the state or properties of an object.

## Core Concepts of OOP

OOP is defined by four core principles, often summarized by the acronym **A PIE**:

1.  **Abstraction:** Hiding the complex implementation details and showing only the essential features of the object to the user. It focuses on *what* an object does rather than *how* it does it.
2.  **Encapsulation:** The mechanism of binding the data and the methods that operate on that data together into a single unit (the **class**). It also restricts direct access to some of an object's components, preventing external interference and misuse.
3.  **Inheritance:** The mechanism where a new class (**derived** or **child** class) is created from an existing class (**base** or **parent** class). This allows the child class to reuse the attributes and methods of the parent class, promoting code reusability. It models the "is-a" relationship.
4.  **Polymorphism:** The ability of an object to take on many forms. Specifically, it allows a single interface or method name to be used for different underlying data types or classes. The two main types are **method overloading** (using the same method name with different parameters) and **method overriding** (redefining a parent class's method in a child class).


## Benefits of OOP

* **Modularity:** Objects are self-contained, making it easier to manage and troubleshoot the software.
* **Reusability:** Inheritance allows for the creation of new classes by reusing existing code, reducing development time.
* **Flexibility (via Polymorphism):** Code can be written more generally to work with a variety of objects.
* **Maintainability:** Encapsulation helps to ensure that changes in one part of the code do not unintentionally affect other parts.
* **Effective Problem Solving:** It allows developers to model real-world concepts directly in code, leading to clearer and more intuitive designs.

## 1. **Encapsulation**

Encapsulation means **bundling data and methods together** and **restricting direct access** to internal data.

```python
class Product:
    def __init__(self, name: str, price: float, stock: int = 0):
        self.name = name
        self.__price = price      # Private attribute
        self.__stock = stock      # Private attribute

    def get_price(self):
        return self.__price

    def set_price(self, new_value):
      self.__price = new_value

    def update_stock(self, quantity):
        if quantity > 0:
            self.__stock += quantity
        else:
            print("Invalid stock quantity")

    def purchase(self, quantity):
        if quantity <= self.__stock:
            self.__stock -= quantity
            print(f"Purchased {quantity} {self.name}(s)")
        else:
            print("Not enough stock!")

# Example
p1 = Product("Laptop", 1500, 5)
p1.purchase(2)
print("Price:", p1.get_price())   # ✅ Access through method
# print(p1.__price)               # ❌ Direct access not allowed
```

🔒 **Encapsulation ensures** that sensitive data (`price`, `stock`) cannot be accessed or modified directly — only through safe methods.

## 🧬 2. **Inheritance**

Inheritance allows a class to **reuse** and **extend** features of another class.

Let’s make `DigitalProduct` and `PhysicalProduct` inherit from `Product`.

```python
class DigitalProduct(Product):
    def __init__(self, name, price, download_link):
        super().__init__(name, price, stock=9999)
        self.download_link = download_link

    def download(self):
        return f"Downloading {self.name} from {self.download_link}"

class PhysicalProduct(Product):
    def __init__(self, name, price, stock, weight):
        super().__init__(name, price, stock)
        self.weight = weight

    def shipping_cost(self, rate_per_kg: float = 5.0) -> float:
        """Calculate shipping cost based on weight."""
        return self.weight * rate_per_kg
```

### Example:

```python
ebook = DigitalProduct("Python eBook", 25, "ebook-download.com")
phone = PhysicalProduct("Smartphone", 800, 10, 0.3)
```

**Inheritance avoids repetition** — we reuse `Product` methods like `purchase()` without rewriting them.



##  3. **Polymorphism**

Polymorphism means **different classes can use the same method name but behave differently**.

Example: different **payment methods** (Credit Card, PayPal, Bank Transfer).

```python
class Payment:
    def pay(self, amount):
        raise NotImplementedError("Subclasses must implement pay()")

class CreditCardPayment(Payment):
    def pay(self, amount):
        print(f"Paid ${amount} using Credit Card.")

class PayPalPayment(Payment):
    def pay(self, amount):
        print(f"Paid ${amount} using PayPal.")

class BankTransferPayment(Payment):
    def pay(self, amount):
        print(f"Paid ${amount} using Bank Transfer.")
```

### Example:

```python
payments = [CreditCardPayment(), PayPalPayment(), BankTransferPayment()]

for method in payments:
    method.pay(100)   # Same method name, different behavior
```

🔄 **Polymorphism** lets us call `.pay()` without caring *which payment type* is used — making the system flexible and easily extensible.



## 4. **Abstraction**

Abstraction hides complex implementation details and exposes only **essential operations**.

Example: `Order` class hides how totals or payments are calculated.

```python
class Order:
    def __init__(self, customer):
        self.customer = customer
        self.items = []
        self.total = 0

    def add_product(self, product, quantity):
        self.items.append((product, quantity))
        self.total += product.get_price() * quantity

    def checkout(self, payment_method):
        print(f"\nProcessing order for {self.customer.name}...")
        payment_method.pay(self.total)
        print(f"Order complete! Total: ${self.total}")
```

The **user of the `Order` class** doesn’t need to know *how* total calculation or payment works — just calls `checkout()`.

## Python Modules and Packages


### Python Module
A **Python module** is a file containing Python code (definitions, functions, classes, and variables), which has the file extension **.py**. It's the simplest way to logically organize related code.



**Why We Need Python Modules**

Modules are essential for building scalable and maintainable programs due to the following reasons:

1.  **Code Reusability:** This is the primary reason. Once you define a function or class in a module, you can use it in any other Python program simply by importing the module. You don't have to rewrite the same code multiple times. For example, the built-in `math` module provides functions like `sqrt()` and `sin()` that you can reuse everywhere.
2.  **Organization and Modularity:** Modules break down large, complex programs into smaller, manageable, and logical units. This makes the code structure clearer, easier to navigate, and simpler to debug. It applies the "divide and conquer" principle to programming.
3.  **Namespace Isolation:** Modules help prevent naming conflicts. If you define a function named `print_report` in your `data_processor.py` module and another one in your `ui_manager.py` module, they don't clash. You access them using the module name: `data_processor.print_report()` and `ui_manager.print_report()`.
4.  **Distribution:** Modules and packages (collections of modules) form the basis for sharing code with the world. Libraries like **NumPy**, **Pandas**, and **Django** are distributed as packages of modules, allowing others to quickly install and leverage powerful functionality.

### Python Packages

A **Python package** is a way of organizing related Python modules into a directory hierarchy. Essentially, a package is a folder that contains multiple module files (`.py` files) and, in older Python versions (prior to Python 3.3), required a special file named `__init__.py` to be recognized as a package.

**Why We Need Python Packages**

Packages are vital for developing large, complex, and professional-grade applications for the following reasons:

1.  **Hierarchical Structuring (Organization):**
    * As a project grows, having hundreds of module files in a single directory becomes unmanageable. Packages allow you to group related modules into subdirectories, creating a logical file system.
    * *Example:*
     An **e-commerce application** might have a top-level package called **`ecommerce`**, with sub-packages logically grouping the core business domains:

      * **`ecommerce.products`**: Contains modules for defining the **Product** class, managing inventory, and handling product catalogs.
      * **`ecommerce.orders`**: Contains modules for the **Order** class, processing checkout, calculating shipping, and updating order status.
      * **`ecommerce.customers`**: Contains modules for the **Customer** class, user authentication, profile management, and handling user accounts.
      * **`ecommerce.payments`**: Contains modules for processing transactions, integrating with different **Payment** gateways (Stripe, PayPal), and handling refunds.

      This structure groups related functionality, making it easy to find where the code for a specific entity, like a **Customer** or an **Order**, resides.

2.  **Namespace Management:**
    * Packages prevent naming collisions across different parts of a large application or library.
    * Since packages define a clear hierarchy, you access modules using their full path (e.g., `import package_name.module_name.function_name`). This ensures that there are no function name collisions, even when different packages contain functions with the same names.

3.  **Code Sharing and Distribution:**
    * Packages are the standard format for distributing large, functional libraries to the wider Python community via the Python Package Index (PyPI).
    * Every major Python library—like **Pandas**, **NumPy**, **Django**, and **Requests**—is structured as a package (or a collection of packages), allowing users to install and use them easily with tools like `pip`.

4.  **Controlled Access and Initialization:**
    * The `__init__.py` file (even when empty) allows you to execute initialization code *when the package is imported*. This can be used to set up package-level variables, define functions that should be immediately available, or control exactly which modules are exposed when a user imports the package.
