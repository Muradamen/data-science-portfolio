# 🍬 The Magic Candy Factory Adventure! 🍭



Today, we're going to visit a **MAGIC CANDY FACTORY**!

The factory makes colorful candies, and we need to help them count and sort the candies. Are you ready? Let's go!

## 📦 Step 1: Opening Our Magic Toolbox

Before we start our adventure, we need to get our magic tools!

Think of these tools like:
- 🧮 A calculator (numpy)
- 📊 A drawing board (matplotlib)
- 📋 A notebook for writing (pandas)

Let's open our toolbox:

```python
# This line brings in numpy - our calculator
# numpy helps us do math
import numpy as np

#This brings in matplotlib - A drawing board
import matplotlib.pyplot as plt

#This brings in pandas - our organizer
#Pandas helps us organize numbers in tables
import pandas as pd

#This is like setting a magic seed so we see all the same candies!
np.random.seed(42)

print("All our magic tools are ready!")
```

## 🍬 Step 2: The Candy Factory Makes Candies!

The candy factory has 10 different rooms, and each room makes candies!

Let's see how many candies each room made today:

```python
# np.random.randint means "pick random whole numbers"
# (5, 15) means  pick numbers between 5 and 15
# size=10 means pick 10 numbers

candies_per_room = np.random.randint(5, 15, size=10)

# Let's name each room: Room1, Room2, etc.
room_names = ['Room' + str(i) for i in range(1, 11)]

# Let's make a table to see our candy
candy_data = pd.DataFrame({
    'Room': room_names,
    'Candies': candies_per_room

})

print("Here's how many candies each room made")
print(candy_data)
```

## 🧮 Step 3: Counting ALL The Candies (Population Mean)

**Population** means EVERYTHING! Like all the toys in your toy box.

Let's count ALL the candies from ALL the rooms and find the average (mean)!

**Average** means: If we shared all candies equally, how many would each room get?

```python
#.mean() is like adding all candies and dividing by number of rooms

all_candies_average = candy_data['Candies'].mean()

print(f" If we look at ALL rooms, each room makes {all_candies_average:.1f} candies on average!")
print(f"\n(This is called the POPULATION MEAN - the average of everything!) ")
```

## 🔍 Step 4: Looking at Just 3 Rooms (Sample Mean)

Sometimes we can't check ALL the rooms. Maybe we're too tired! 😴

So we pick just a FEW rooms to check. This is called a **SAMPLE**!

Let's pick 3 rooms and see their average:

```python
# .head() means  "give me the first 5 rooms"
sample_rooms = candy_data.head(5)

# calculate average for just these three rooms
sample_average = sample_rooms['Candies'].mean()

print("We checked 5 rooms")
print(sample_rooms)
print(f"\n These five rooms make  {sample_average:.1f} candies on average ")
print(f"\n (This is called the SAMPLE MEAN - average of just a few) ")
```

## 📊 Step 5: Drawing a Picture of Our Candies!

Let's draw a colorful picture to see how many candies each room made!

Each bar will be like a tower of candies! 🏗️

```python
# plt.figure makes a space for or drawing
# figsize=(10, 6) means make it this big
plt.figure(figure=(10, 6))

# plt.bar draws tall bars
# color='skyblue' makes them pretty blue
plt.bar(candy_data['Room'], candy_data['Candies'], color='skyblue', edgecolor='navy')

# Add labels so we know what we are looking at
plt.xlabel('Factory Rooms', fontsize=12)
plt.ylabel('Number of Candies', fontsize=12)
plt.title('Candies Made by room', fontsize=14,fontweight='bold')

# Rotate room names

plt.xticks(rotation=45)

# Show our beautiful picture
plt.tight_layout()
plt.show()
```

## 🎲 Step 6: The Difference Game (Variance)

**Variance** is a big word that means: "How different are the numbers?"

Imagine:
- If all rooms made 10 candies → Very similar! Low variance! 😊
- If one room made 5 and another made 15 → Very different! High variance! 😮

Let's check how different our rooms are:

```python
#.var()tells us how spread out the numbers are
# ddof=0 means wer're looking at all rooms(population)
all_rooms_variance = candy_data['Candies'].var(ddof=0)

# ddof=1 means we're looking at just some rooms (sample)
sample_variance = sample_rooms['Candies'].var(ddof=1)

print(f" ALL rooms variance: {all_rooms_variance:.2f}")
print(f" just 3 rooms variance: {sample_variance:.2f}")
print(f"\n Higher number = rooms make very different amount of candies")
print(f"\n lower number = rooms make SIMILAR amount of candies")

```

## 🎪 Step 7: The Magic Candy Bell! (Normal Distribution)

There's a magic rule in candy factories: When we make LOTS of candies, they usually form a **BELL SHAPE**! 🔔

Most candies are in the middle, and fewer are very small or very big.

Let's imagine a HUGE candy factory with 1000 pieces!

```python
# Let's make 1000 candy sizes
# They'll be around size 10, give or take about 2

many_candies = np.random.normal(loc=10, scale=2, size=1000)

# Draw a picture of all these candy sizes
plt.figure(figsize=(10, 6))

# plt.hist makes a histogram
# bins=30 means divide into 30 groups
plt.hist(many_candies, bins=30, color='pink', edgecolor='red', alpha=0.7)

plt.xlabel('Candy Size', fontsize=12)
plt.ylabel('How many Candies', fontsize=12)
plt.title('The magic Candy Bell', fontsize=14, fontweight='bold')

plt.show()


```

## 🎲 Step 8: The Magic Dice Game! (Probability)

**Probability** is a fancy word for "How likely is something to happen?"

Let's play with a magic candy die! 🎲

Imagine a die with 6 sides, each showing a different candy color:
- 🔴 Red
- 🔵 Blue  
- 🟢 Green
- 🟡 Yellow
- 🟣 Purple
- 🟠 Orange

What's the chance of getting a red candy?

```python
# use special math tool called scipy
import scipy.stats as stats

# Take our sample of 3 rooms
sample_candies = sample_rooms['Candies']

# Calculate the averge
mean = np.mean(sample_candies)

# calculate (standard error)/ how much the numbers wiggle around
standard_error = stats.sem(sample_candies)

# Count rooms we checked, minus 1
degrees_freedom = len(sample_candies) -1

# Make our confident guess! (95% confident)
confidence_interval = stats.t.interval(0.95,degrees_freedom, mean,standard_error)


print(f"Based on 3 rooms, we're 95% sure that...")
print(f"The REAL average for all rooms is between:")
print(f" {confidence_interval[0]:.1f} and {confidence_interval[1]:.1f} candies!")
print(f"\n The real average we found earlier was: {all_candies_average:.1f}")
print(f"  Is it between our guessess? let's check! ")

if confidence_interval[0] <= all_candies_average <= confidence_interval[1]:
    print(" yes! our guess was right! we are too smart give us our certificate now!")
else:
    print(" pride cometh before a fall we are sorry Kayode we love you. You are the greatest! the best! the best of the best!")
```
