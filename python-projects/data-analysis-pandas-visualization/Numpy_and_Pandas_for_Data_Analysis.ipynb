# NumPy
NumPy is a Python library that supports numerical computing. It is an open source project created by Travis Oliphant in 2005. It makes working with multi-dimensional arrays easy.

It was created due to the unique challenges posed by numerical computations, such as those encountered in fields like data science, machine learning, physics, and engineering.

Since NumPy arrays are implemented in C, they are optimized for performance, providing faster execution of mathematical operations compared to Python lists. And unlike Python lists, NumPy arrays tolerate homogeneous data types, which are stored in a contiguous block of memory, leading to efficient storage and access.

## Creating NumPy Array
A NumPy array can consist of one or more dimensions, and it can be created from a Python list. It can also be synthesized from scratch using functions like `np.random.random`, `np.zeros`, etc.

```python
# Import NumPy
import numpy as np
```

### 1D NumPy array (Vector)

A **1-dimensional (1D) NumPy array** is a simple linear collection of elements stored under a single axis. It is similar to a Python list but provides faster computation and supports advanced mathematical operations.

#### Key Features

* Has only **one axis** (axis = 0).
* Elements are stored in a continuous block of memory → efficient processing.
* Supports vectorized operations (e.g., add, multiply) without loops.
* All elements typically share the **same data type** (`dtype`).

#### Example

* Can be visualized as a **cube** or multiple matrices layered on top of each other
* Efficient storage and manipulation of multi-layered numeric data

## **Reshaping arrays**

**Array reshaping** is the process of changing the dimensions (shape) of a NumPy array **without changing its data**. For instance we can convert a 1D array into 2D, or 2D into 3D, etc.



### Why do we reshape arrays?

1. **To meet model or function requirements**: Machine learning models often expect data in specific shapes
(e.g., samples × features).

2. **To organize data meaningfully**: Reshaping helps represent data as:

    * 2D tables
    * Images (2D or 3D)
    * Batches for training models

3. **For matrix operations**: Some linear algebra operations require matrices instead of vectors.

4. **To prepare input/output**: Data from files may come in 1D; reshaping prepares it for use.

5. **Memory efficiency**: It rearranges structure without copying data (fast and lightweight).



```python
arr = np.array([1, 2, 3, 4, 5, 6])
arr
```

```python
reshaped = arr.reshape(2, 3)   # 2 rows, 3 columns
print(reshaped)
```

### Vectorised operation
**Broadcasting**

Adding a scalar to all list elements

```python
# Create list of lists
list_of_lists = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
list_of_lists
```

```python
list_of_lists + 3
```

```python
# Create an empty list to store the result
result = []

# Iterate through the outer list
for inner_list in list_of_lists:
    # Create an empty list for each inner list in the result
    new_inner_list = []

    # Iterate through the elements of the inner list
    for element in inner_list:
        # Add 3 to each element and append to the new_inner_list
        new_inner_list.append(element + 3)

    # Append the new_inner_list to the result
    result.append(new_inner_list)

```

```python
vectorised_opp_array = np.array(list_of_lists)
```

```python
vectorised_opp_array + 3
```

```python
# View the results
result
```

```python
array_8 = np.array(list_of_lists)
array_8
```

```python
array_8 + 3
```

**Create a function to pull sales records**

```python

def filter_sales_records(df, salesperson_name, start_date, end_date):
    """
    Filter sales records based on salesperson name and date range.

    Parameters:
    - df: DataFrame with columns 'Date', 'Salesperson', 'Total Sales'
    - salesperson_name: Name of the salesperson to filter records for
    - start_date: Start date of the desired date range
    - end_date: End date of the desired date range

    Returns:
    - DataFrame containing filtered sales records
    """
    # Filter based on salesperson name
    filtered_df = df[df['salesperson'] == salesperson_name]

    # Filter based on date range
    date_mask = (filtered_df['Date'] >= start_date) & (filtered_df['Date'] <= end_date)
    filtered_df = filtered_df[date_mask]

    return filtered_df
```
